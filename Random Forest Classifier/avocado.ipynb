#### Import necessary libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier

import warnings

# Suppress Matplotlib warnings
warnings.filterwarnings('ignore', category=UserWarning)
```

#### Analyses of given DataSet

```python
dataFrame = pd.read_csv('avocado.csv')
dataFrame.info()
dataFrame.isnull().sum()
```

```python
dataFrame.describe().transpose().applymap(lambda x: format(x, '.3f'))
```

```python
dataFrame = dataFrame.drop('Unnamed: 0', axis=1)
```

```python
dataFrame
```

```python
non_numeric_columns = ['Date', 'type', 'region']

# Convert non numeric columns
label_encoder = LabelEncoder()
dataFrame[non_numeric_columns] = dataFrame[non_numeric_columns].apply(lambda col: label_encoder.fit_transform(col))

# Calculate the correlation matrix
correlation_matrix = dataFrame.corr()

# Create a heatmap using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Map')
plt.show()
```

#### Decision Tree 

```python
# Calculate correlation coefficients
correlation_matrix = dataFrame.corr()
correlation_with_target = correlation_matrix['AveragePrice'].abs().sort_values(ascending=False)
```

```python
# Choose the top two features with the highest absolute correlation
top_features = correlation_with_target[1:3].index
```

```python
# Create the decision boundary graph
sns.pairplot(dataFrame, x_vars=top_features, y_vars='AveragePrice', hue='type', markers=["o", "s"])
plt.show()
```

Тут видно что органические авокадо стоят дороже а не органические дешевле,а также то что вид авокадо под номером 4046 весь не органический и его цена средняя относительно других видов авокадо

```python
X=dataFrame[['type', '4046']]
y=dataFrame[['AveragePrice']]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)

XX=dataFrame[['AveragePrice', '4046']]
YY=dataFrame[['type']]
XX_train,XX_test,yy_train,yy_test=train_test_split(XX,YY,test_size=0.3,random_state=42)
```

```python
# Hyperparameters to explore
hyperparameters = {
    'max_depth': np.arange(1, 100),
    'min_samples_split': np.arange(2, 100),
    'min_samples_leaf': np.arange(1, 100),
    'min_weight_fraction_leaf': np.linspace(0, 0.5, 11),
    'max_features': np.arange(2,100),
    'max_leaf_nodes': np.arange(2, 100),
    'min_impurity_decrease': np.arange(0.01, 100)
}
```

```python
# Plotting
plt.figure(figsize=(15, 10))

for hyperparameter, values in hyperparameters.items():
    accuracies = []

    for value in values:
        # Create a DecisionTreeClassifier with the specific hyperparameter value
        clf = DecisionTreeRegressor(**{hyperparameter: value})
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        accuracy = metrics.mean_squared_error(y_test, y_pred)
        accuracies.append(accuracy)

    # Plotting the line
    plt.plot(values, accuracies, label=hyperparameter)

plt.xlabel('Hyperparameter Values')
plt.ylabel('Accuracy')
plt.title('Influence of Hyperparameters on Model Accuracy')
plt.legend()
plt.show()
```

```python
best_model_DecisionTreeRegressor = DecisionTreeRegressor(max_depth=70,
                                                         min_samples_split=2,
                                                         min_samples_leaf=1,
                                                         min_weight_fraction_leaf=0,
                                                         max_features=2,
                                                         max_leaf_nodes=2,
                                                         min_impurity_decrease=20,
                                                         random_state=42)
best_model_DecisionTreeRegressor.fit(X_train, y_train)
best_model_DecisionTreeClassifier = DecisionTreeClassifier()
best_model_DecisionTreeClassifier.fit(XX_train, yy_train)
```

```python
plot_tree(best_model_DecisionTreeRegressor, filled=True, rounded=True, fontsize=10)
plt.show()

plt.figure(figsize=(10, 25))
plot_tree(best_model_DecisionTreeClassifier, filled=True, rounded=True, fontsize=10)
plt.show()
```

#### Random Forest Classifier

```python
# Standardize the features
scaler = StandardScaler()
XX_train_scaled = scaler.fit_transform(XX_train)
XX_test_scaled = scaler.transform(XX_test)
```

```python
# Create and fit a Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(XX_train_scaled, yy_train.values.ravel())
```

```python
# Plot decision boundary
plt.figure(figsize=(10, 8))

# Plot the decision boundary
h = .02  # step size in the mesh
x_min, x_max = XX_train_scaled[:, 0].min() - 1, XX_train_scaled[:, 0].max() + 1
y_min, y_max = XX_train_scaled[:, 1].min() - 1, XX_train_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = rf_model.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.3)
```

Из таблицы видно что органические и не органические авокадо достаточно ровно разделены друг от друга
если взять два параметра 4046 и Среднюю цену за авокадо

```python
# Plot the training points
plt.scatter(XX_train_scaled[:, 0], XX_train_scaled[:, 1], c=yy_train.values.ravel(), cmap=plt.cm.coolwarm, edgecolors='k', marker='o', s=100, label='Training points')

# Plot the testing points
plt.scatter(XX_test_scaled[:, 0], XX_test_scaled[:, 1], c=yy_test.values.ravel(), cmap=plt.cm.coolwarm, edgecolors='k', marker='x', s=100, label='Testing points')

plt.xlabel('AveragePrice')
plt.ylabel('4046')
plt.title('Decision Boundary - Random Forest')
plt.legend()
plt.show()
```

Из данной таблицы видно как тестовые и тренировочные данные распределены если мы возьмем в качестве параметров 4046 и Среднюю за авокадо
как видно органические и не органические авокадо тоже хорошо распределены

```python
#Train Model
best_rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=15,
    random_state=42
)

best_rf_model.fit(XX_train, yy_train.values.ravel())

# Make predictions on the test set
y_pred = best_rf_model.predict(XX_test)

# Evaluate the model
accuracy = accuracy_score(yy_test, y_pred)
print(f"Accuracy on the test set: {accuracy:.2%}")
```
